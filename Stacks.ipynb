```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        
    def Push(self,data):
        if self.top is None:
            self.top = Node(data)
            return
        temp = Node(data)
        temp.next = self.top
        self.top = temp
        temp = None
    
    def Pop(self):
        if self.top is None:
            print("List is Empty\n")
            return
        key = self.top.data
        temp = self.top
        self.top = self.top.next
        temp.data = None
        temp.next= None
        temp = None
        return key
        
    def Is_Empty(self):
        if self.top is None:
            return True
        else:
            return False
    
    def Peek(self):
        if self.top == None:
            return None
        print(self.top.data)
        return self.top.data
    
    def traverse(self):
        move = self.top
        
        while move is not None:
            print(move.data, end= ' -> ')
            move = move.next
            
        print('None') 
    
    def Queue(self):
        temp_stack = Stack()
        while self.top:
            data = self.Pop()
            temp_stack.Push(data)
        return temp_stack
    
    
```
