```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.top = None
        
    def Push(self,data):
        if self.top is None:
            self.top = Node(data)
            return
        temp = Node(data)
        temp.next = self.top
        self.top = temp
        temp = None
    
    def Pop(self):
        if self.top is None:
            print("List is Empty\n")
            return
        key = self.top.data
        temp = self.top
        self.top = self.top.next
        temp.data = None
        temp.next= None
        temp = None
        return key
        
    def Is_Empty(self):
        if self.top is None:
            return True
        else:
            return False
    
    def Peek(self):
        if self.top == None:
            return None
        print(self.top.data)
        return self.top.data
    
    def traverse(self):
        move = self.top
        
        while move is not None:
            print(move.data, end= ' -> ')
            move = move.next
            
        print('None') 

    def trav(self):
        move = self.top
        while move:
            move2 = move.data.top
            while move2:
                print(move2.data, end=" -> ")
                move2 = move2.next
            move = move.next 
                
        print("None")
        
    def PopIndex(index):
        move1 = self.top
        l = 0
        
        while move1:
            move1 = move1.next
            l += 1
            
        move1 = self.top
        cou = 0
        while cou < l - index:
            move1 = move1.next
            cou += 1
        move1 = move1.data
        cou = move1.Pop()
        return cou
        
```
